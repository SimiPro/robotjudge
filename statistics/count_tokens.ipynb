```python
#####################
#     LOAD DATA     #
#####################

import json_lines
import csv

def process_tweet(tweet):  
    d = {}
    d['hashtags'] = [hashtag['text'] for hashtag in tweet['entities']['hashtags']]
    d['text'] = tweet['full_text']
    d['user'] = tweet['user']['screen_name']
    d['user_loc'] = tweet['user']['location']
    d['created_at'] = tweet['created_at']
    return d

                    
import pandas as pd
tweets = pd.read_csv("../senators-1-tweets.csv", header=None, names=['hashtags', 'text', 'user', 'user_location', 'created_at'])  
print('num tweets: {}'.format(len(tweets)))


import spacy
nlp = spacy.load('en_core_web_md')

import random
        
```

```python
tweets_tbl = tweets.values
```

```python
# make docs
docs = []
N = tweets_tbl.shape[0]
for i in range(N):
    if i % 1000 == 0:
        print('{}%'.format(100./N*i), end=' ')
    docs.append(nlp(tweets_tbl[i][1]))
```

```python
# count tokens
tokens = set()
for doc in docs:
    for t in doc:
        tokens.add(t)
print(f'number of tokens: {len(tokens)}')
```
